```python
tokenizer = RegexpTokenizer('[A-Za-z]+')
stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()

def clean_text(row: pd.Series) -> str:
    """Function applied to text column in each row.
    
    Args:
        row(pandas.Series):   input rows as a series.
    
    Returns:
        str:    output string of tokens after cleaning.
    """
    
    # Get the contents of the text field.
    str_ = row.text
    
    # Remove all numbers and punctuations.
    list_ = tokenizer.tokenize(str_)

    # Change all words to lowercase.
    list_ = [w.lower() for w in list_]
    
    # Remove all stop words.
    list_ = [w for w in list_ if w not in stop_words]
    
    # Stem all words.
    list_ = [stemmer.stem(w) for w in list_]

    # How should results be packaged? 
    # - As one string of tokens per row.
    return " ".join(list_)


# Note: we need X to be a dataframe not a series!
X_processed = X.apply(clean_text, axis=1).to_frame()
print(X_processed.shape)
print(X_processed.head())
print(X.head())

# midterm review questions:
# - what is 'text' in the line: str_ = row.text ?
# - after the code runs, does X still have the original strings?
```
