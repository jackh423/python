1. From `countries.xlsx` file, read in all data into a DataFrame, make the country names the row labels (the index for the rows), and remove the SpecialNotes column.
<br>a. Print the number of countries in the resulting DataFrame
<br>b. Print the first 3 lines of country data in the resulting DataFrame.

```python
# Reading all data from countries.xlsx into a data frameand making "CountryName" as row index
country_df = pd.read_excel("countries.xlsx", index_col='CountryName')
# print(country_df)
print(type(country_df))

# Dropping the "SpecialNotes" column 
country_df.drop(columns=['SpecialNotes'], inplace=True)
print(country_df)
print()

# Printing number of countries in the dataframe
print(f"Number of Countries: {len(country_df)}")

# Printing the shape of data frame. Just for practice
print(country_df.shape)

# Printing first 3 lines of country data
print(country_df.iloc[:3], "\n")
print("------------------------------------------------------------------------")

# Alternative way of printing first 3 rows
print(country_df.head(3))
print("------------------------------------------------------------------------")

# 3rd way of printing first 3 rows of data
print(country_df.loc["Aruba":"Angola"], "\n")
```

2. Shorten the column names so it's easy to type them:
<br>CountryCode => Code, Region => Region, LevelOfIncome => Income
<br>Challenge: do this in one line of code
<br>Print the first lines of country data in the DataFrame

```python
# Printing country_df before renaming
print(country_df)
print("------------------------------------------------------------------------")

# renaming the column names. This won't do renaming inplace instead it returns a new df
df = country_df.rename(columns={"CountryCode": "Code", "Region": "Region", "LevelOfIncome": "Income"})

# Printing the df after renaming the columns
print(df)
print("------------------------------------------------------------------------")

# renaming the column names inplace. So no new df is returned
country_df.rename(columns={"CountryCode": "Code", "Region": "Region", "LevelOfIncome": "Income"}, inplace=True)

# Printing the df after renaming the columns
print(country_df)

print("------------------------------------------------------------------------")
```

3. From `population.xlsx` file, read in all data into another DataFrame, and also make the country names the row labels.
<br>Remove the Country Code, Indicator Name, Indicator Code, and 2020 columns since they contain duplicate data or unnecessary data.
<br>a. Print the number of rows and columns of the resulting DataFrame.
<br>b. Print the last 3 lines of the Dataframe

```python
# Reading data into df from "population.xlsx"
pop_df = pd.read_excel("population.xlsx", index_col='Country Name')
# print(pop_df)

# Droping Country Code, Indicator Name, Indicator Code, and 2020 columns 
pop_df.drop(columns=['Country Code', "Indicator Name", "Indicator Code", "2020"], inplace=True)

# Print the number of rows and columns of the resulting DataFrame
print(f"Number of rows and columns of the DF: {pop_df.shape}")

# Print the last 3 lines of the Dataframe
print(pop_df.tail(3))


# Just for practice
# print(pop_df)
# print(pop_df.index)
# print(pop_df.columns)
```

4. Combine the 2 DataFrames side by side into one DataFrame that has:
<br>- the same number of rows as the 2 original DataFrames
<br>- the country names are the row labels
<br>- the order of data for each country is: Code, Region, Income, then all the years.
<br>Print the first 5 lines of data the resulting DataFrame.

```python
combined_df = pd.concat([country_df, pop_df], axis=1)
# print(combined_df)
print(combined_df.head(5))
```

6a. Sometimes we can work with exact data when doing data analysis.
<br>
<br>Find and print the 2019 total population of each region of the world, sorted by the population in descending order (highest population first).
<br>To print the population values with comma, the pandas float_format has been set for you.

```python
pd.options.display.float_format = '{:,.0f}'.format

# Creating df with Region and 2019 population
pop_2019_df = combined_df[["Region", "2019"]]
# print(pop_2019_df)

# Printing population of each region of the world in descending order
per_reg_df = pop_2019_df.groupby("Region").sum().sort_values(by="2019", ascending=False)
# print(pop_2019_df.groupby("Region").sum().sort_values(by="2019", ascending=False), "\n")
print(type(per_reg_df))
print(per_reg_df)
```

6b. Plot the 2019 population of each region. Use the appropriate plot, and be sure to have the title, x-axis label, y-axis label, and appropriate ticks.
<br>From the plot you can see which parts of the world have the most or the least people.

```python
import matplotlib.pyplot as plt
%matplotlib inline
# For this problem, I have drawn 3 different plots (Line, Bar (2 charts), and Pie) to show the same data in different formats
# That way, I could able to show different views of the data for better visualtion comparision.
# Personally, I think Pie chart is representing this data in a better way for visualization.

# 1 Line Chart
# print(per_reg_df.index)
plt.figure(figsize=(16,8))
plt.style.use('seaborn-whitegrid')
plt.plot(per_reg_df.index,per_reg_df[['2019']], color='green')
plt.xticks(rotation=90)
plt.xlabel("Region",fontsize=12)
plt.ylabel("2019 Population",fontsize=12)
plt.title("2019 Total Population Per Region",fontsize=16)
plt.show()

print("#####################################################################")
# Bar Chart
ax = per_reg_df[['2019']].plot(kind='bar', title ="2019 Total Population Per Region", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Region", fontsize=12)
ax.set_ylabel("2019 Population", fontsize=12)
plt.show()

print("#####################################################################")
# Bar Chart
# print(len(per_reg_df))
plt.figure(figsize=(18,8))
# plt.bar(np.arange(len(per_reg_df)), per_reg_df[['2019']], align="center")
plt.bar(np.arange(len(per_reg_df)), per_reg_df["2019"], align="center")
plt.xticks(np.arange(len(per_reg_df)),per_reg_df.index.values) 
plt.xticks(rotation=75)
plt.xlabel("Region",fontsize=12)
plt.ylabel("2019 Population",fontsize=12)
plt.title("2019 Total Population Per Region",fontsize=16)
# plt.xlabel("Region")
# plt.ylabel("2019 Population")

plt.show()


print("#####################################################################")
plt.figure(figsize=(20,12))
# plot chart
ax1 = plt.subplot(111, aspect='equal')
ax1.set_title("2019 Total Population Per Region")
per_reg_df.plot(kind="pie", y ="2019", ax=ax1, autopct='%1.1f%%', startangle=90, shadow=False,)
# df.whatever.value_counts().plot(kind='pie')
plt.legend(loc="upper left", fontsize=10 )
plt.show()
```

<br>Plot the population of all the regions in the past 40 years, in order to see their growth rate.

<br>For the years, print out every 10 years: 1960, 1970, ..., 2000, 2010, 2020

```python
p_df = combined_df.groupby("Region").sum()
# print(p_df)
years = p_df.columns
# print(years)
# print(years)
plt.figure(figsize=(16,8))
plt.style.use('seaborn-whitegrid')
# plt.plot(p_df.index, p_df, color='green', label="Region")
plt.plot(years, p_df.loc['East Asia & Pacific'], label="East Asia & Pacific")
plt.plot(years, p_df.loc['Europe & Central Asia'],  label="Europe & Central Asia")
plt.plot(years, p_df.loc['Latin America & Caribbean'],  label="Latin America & Caribbean")
plt.plot(years, p_df.loc['Middle East & North Africa'],  label="Middle East & North Africa")
plt.plot(years, p_df.loc['North America'],  label="North America")
plt.plot(years, p_df.loc['South Asia'],  label="South Asia")
plt.plot(years, p_df.loc['Sub-Saharan Africa'],  label="Sub-Saharan Africa")

# plt.plot(p_df.columns, p_df, color='green')
plt.xticks(rotation=60)
plt.xlabel("Year",fontsize=12)
plt.ylabel("Population",fontsize=12)
plt.title("Population Growth Rate per Region",fontsize=18)
xaxis = list(years.values)
xaxis.append("2020")
plt.xticks(xaxis[::10])

plt.legend()
# plt.margins(x=0)
plt.show()

# ax.xaxis.set_ticks(x[::10])
# plt.xticks(x[0], x[-1])
```
